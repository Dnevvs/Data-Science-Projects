## Описание проекта

### Цель проекта

Цель данного проекта - разработка двух моделей прогнозирования  на основе предоставленных заказчиком данных с целью оптимизации управления кадровыми ресурсами. Первая модель предназначена для прогнозирования уровня удовлетворенности сотрудника, вторая - направлена на предсказание возможного увольнения сотрудника. Данное решение позволит заказчику более эффективно осуществлять управление персоналом и предупреждать возможный отток кадров.

### Постановка задачи

Компания «Работа с заботой» предоставила данные с характеристиками сотрудников компании. Среди них — уровень удовлетворённости сотрудника работой в компании. Эту информацию получили из форм обратной связи: сотрудники заполняют тест-опросник, и по его результатам рассчитывается доля их удовлетворённости от 0 до 1, где 0 — совершенно неудовлетворён, 1 — полностью удовлетворён. 
У нас несколько задач: 
1. Построить модель, которая сможет предсказать уровень удовлетворённости сотрудника на основе данных заказчика. 
2. Построить модель, которая сможет на основе данных заказчика предсказать то, что сотрудник уволится из компании.

### Описание данных

**Таблицы:**

_**Задача 1:**_
- `train_job_satisfaction_rate.csv` - тренировочная выборка.
- `test_features.csv` - входные признаки тестовой выборки.
- `test_target_job_satisfaction_rate.csv` - целевой признак тестовой выборки.

_**Задача 2:**_
- `train_quit.csv` - тренировочная выборка.
- `test_features.csv`  - входные признаки тестовой выборки (те же, что и в 1-й задаче).
- `test_target_quit.csv` - целевой признак тестовой выборки.

**Признаки:**
- `id` — уникальный идентификатор сотрудника.
- `dept` — отдел, в котором работает сотрудник.
- `level` — уровень занимаемой должности.
- `workload` — уровень загруженности сотрудника.
- `employment_years` — длительность работы в компании (в годах).
- `last_year_promo` — показывает, было ли повышение за последний год.
- `last_year_violations` — показывает, нарушал ли сотрудник трудовой договор за последний год.
- `supervisor_evaluation` — оценка качества работы сотрудника, которую дал руководитель.
- `salary` — ежемесячная зарплата сотрудника.
- `job_satisfaction_rate` — уровень удовлетворённости сотрудника работой в компании, целевой признак.
- `quit` — статус сотрудника(уволился/не уволился).

````python
# Функция применения методов к датасету
def df_research_method(df, method):
    """
    Применяет определенный метод изучения данных в датасете.

    Параметры:
    - df: Исследуемый датасет.
    - method: Название метода.
   
    Пример:
    ```
    df_research_method(df, 'head')
    ```
    """
    
    if method == 'head':
        display(df.head())
    elif method == 'shape':
        display(df.shape)
    elif method == 'describe':
        display(df.describe().T)
    elif method == 'duplicated':
        dp = df.duplicated().sum()
        display(dp)
        if dp > 0:
            display(df[df.duplicated()])
    elif method == 'unique':
        for column in df.select_dtypes(include='object').columns:
            print(f'Уникальные значения в колонке {column}: {df[column].unique()}')
    elif method == 'msno':
        display(msno.bar(df))
        plt.show()
    else:
        display(df.info())

# Функция применения методов к списку датасетов
def df_research(df_list, df_names, df_methods):
    """
    Применяет метод из списка для изучения данных в списке датасетов.

    Параметры:
    - df_list: Список исследуемых датасетов.
    - df_name: Список название исследуемых датасетов.
    - df_method: Список названий методов.
   
    Пример:
    ```
    df_research(df_list, df_names, df_methods)
    ```
    """
    
    for method in df_methods:
        print(f'Изучим {method[1]}:')
        print('-' * 70)
        i = 0
        for df in df_list:
            df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
            print('Dataset:', df_names[i])
            print('-' * 70)
            df_research_method(df, method[0])
            print('-' * 70)
            i += 1;
````

````python
# Функция для отображения категориального признака    
def show_categorical_feature(df, column, title, x_label, y_label):
    """
    Отображает столбчатую и круговую диаграммы для категориального признака.

    Параметры:
    - df: Датасет.
    - column: Название колонки признака.
    - title: Заголово диаграммы.
    - x_label: Название оси X.
    - y_label: Название оси Y.
   
    Пример:
    ```
    show_categorical_feature(market_file_df, 
                         'покупательская_активность',
                         'класс покупательской активности',
                         'Покупательская активность',
                         'Количество покупателей')
    ```
    """
    
    cat_df = df[column].value_counts()
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))
    plt.suptitle(title.upper(), fontsize=16, ha='center')
    sns.barplot(x=cat_df.index,
                y=cat_df,
                color='g',
                ax=axes[0]) 
    axes[0].set_xlabel(x_label)
    axes[0].set_ylabel(y_label)
    axes[0].tick_params(labelrotation=0)
    axes[0].bar_label(axes[0].containers[0],
                      padding=-25,
                      color='white',
                      fontsize=12,
                      label_type='edge',
                      fontweight='bold')
    plt.pie(cat_df, 
            labels=cat_df.index, 
            autopct='%1.1f%%', 
            wedgeprops = { 'linewidth': 3, 'edgecolor': 'white' })
    axes[1].set_xlabel(x_label)
    plt.show()
    
# Функция для отображения количественного признака    
def show_numeric_feature(df, column, title, x_label, y_label):
    """
    Отображает гистограмму и "ящик с усами" для количественного признака.

    Параметры:
    - df: Датасет.
    - column: Название колонки признака.
    - title: Заголовок диаграммы.
    - x_label: Название оси X.
    - y_label: Название оси Y.
   
    Пример:
    ```
    show_numeric_feature(market_file_df, 
                         'выручка',
                         'Выручка за период',
                         'Выручка',
                         'Количество покупателей')
    ```
    """
    
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))
    plt.suptitle(title.upper(), fontsize=16, ha='center')
    sns.histplot(data=df[column], color='g', kde=False, ax=axes[0]) 
    axes[0].set_xlabel(x_label)
    axes[0].set_ylabel(y_label)
    sns.boxplot(data=df[column], color='c')
    axes[1].set_ylabel(x_label)
    plt.show()
    print('Описание признака:', x_label)
    display(df[column].describe().T)
    

# Функция для отображения количественного признака по списку
def show_numeric_feature_by_list(df, features):
    """
    Отображает гистограмму и "ящик с усами" для количественного признака
    по списку.

    Параметры:
    - df: Датасет.
    - features: Список признаков для отображения.
   
    Пример:
    ```
    show_numeric_feature_by_list(market_file_df, 
        market_file_numeric_features)
    ```
    """
    
    for feature in features:
        title = feature[2]
        column = feature[0]
        x_label = feature[1]
        y_label = 'Количество сотрудников'
        show_numeric_feature(df, column, title, x_label, y_label)
    
# Функция для отображения категориального признака по списку
def show_categorical_feature_by_list(df, features):
    """
    Отображает столбчатую и круговую диаграммы для категориального признака
    по списку

    Параметры:
    - df: Датасет.
    - features: Список признаков для отображения.
   
    Пример:
    ```
    show_categorical_feature_by_list(market_file_df, 
        market_file_categorical_features)
    ```
    """

    for feature in features:
        title = feature[2]
        column = feature[0]
        x_label = feature[1]
        y_label = 'Количество сотрудников' 
        show_categorical_feature(df, column, title, x_label, y_label)

# Функция для отображения количественного признака в разрезе целевого
def show_numeric_feature_by_target(df, target_column, features):
    """
    Отображает гистограмму и "ящик с усами" для количественного признака
    в разрезе целевого признака.

    Параметры:
    - df: Датасет.
    - target_column: Название колонки целевого признака.
    - features: Список признаков для отображения.
   
    Пример:
    ```
    show_numeric_feature_by_target(market_file_df, 
        'выручка',
        market_file_numeric_features)
    ```
    """
    colors = 'ygb'
    quits=['Не уволился', 'Уволился']
    targets = sorted(df[target_column].unique())
    for feature in features:
        title = feature[2]
        column = feature[0]
        x_label = feature[1]
        y_label = 'Количество сотрудников'
        # Построим гистограммы
        plt.figure(figsize=[18, 12])
        plt.subplots_adjust(top=0.95, wspace=0.2, hspace=0.3)
        plt.suptitle(f'Диаграмма распределения по целевому признаку - {title}', fontsize=15)
        plt.subplot(len(targets), 1, 1)
        i = 0
        for target in targets:
            target_df = df.query(f'{target_column} == @target')
            sns.histplot(target_df[column], kde=False, label=target, color=colors[i])
            i += 1
        plt.xlabel(f'{quits[1]}')                    
        plt.ylabel(y_label)
        plt.legend(bbox_to_anchor=(0.65, -0.1),  ncol=len(targets))
        plt.show()
        # Построим ящики с усами
        plt.figure(figsize=[18, 6])
        plt.subplots_adjust(top=0.95, wspace=0.2, hspace=0.3)
        sns.boxplot(x=target_column, y=column, data=df) 
        plt.ylabel(x_label)
        plt.xlabel(f'{quits[1]}')
        plt.show()
        # Выведем информацию по таргетам
        i = 0
        for target in targets:
            target_df = df.query(f'{target_column} == @target')
            print(f'{quits[i]}')
            print('-' * 70)
            display(target_df[column].describe())
            print(f'Итого по таргету: {target_df[column].sum()}')
            print('-' * 70)
            i += 1

# Функция для отображения категориального признака в разрезе целевого
def show_categorical_feature_by_target(df, target_column, features, in_percents=True):
    """
    Отображает столбчатую и круговую диаграммы для категориального признака
    в разрезе целевого признака.

    Параметры:
    - df: Датасет.
    - target_column: Название колонки целевого признака.
    - features: Список признаков для отображения.
    - percents: показывать в процентах (по умолчанию True).
   
    Пример:
    ```
    show_categorical_feature_by_target(market_file_df, 
        'покупательская_активность',
        market_file_categorical_features, in_percents=False)
    ```
    """

    colors='gr'
    quits=['Не уволился', 'Уволился']
    targets = sorted(df[target_column].unique())
    for feature in features:
        title = feature[2]
        column = feature[0]
        x_label = feature[1]
        y_label = 'Количество сотрудников' 
        fig, axes = plt.subplots(1, len(targets), figsize=(18, 8))
        plt.suptitle(title.upper(), fontsize=16, ha='center')
        i = 0
        for target in targets:
            target_df = df.query(f'{target_column} == @target')[column].value_counts(normalize=True)
            if in_percents:
                percents = (target_df / target_df.sum()) * 100 
                dx = zip(target_df.index, percents)
                target_df = pd.DataFrame(dx, columns=[column, 'percent']).sort_values(
                    by='percent', ascending=False)
                sns.barplot(x=target_df[column].astype(str), 
                            y=target_df['percent'],
                            color=colors[i],
                            ax=axes[i]) 
                axes[i].set_xticks(range(len(percents)))
                axes[i].set_xticklabels(target_df[column].to_list())
            else:
                sns.barplot(x=target_df.index, 
                            y=target_df,
                            color=colors[i],
                            ax=axes[i]) 
            axes[i].set_title(x_label)
            axes[i].set_xlabel(f'{quits[i]}')
            axes[i].set_ylabel(y_label)
            axes[i].tick_params(labelrotation=0)
            if in_percents:
                axes[i].bar_label(axes[i].containers[0], 
                              labels=target_df['percent'].round(1).astype(str) + '%',
                              fmt='%.1i',
                              color=colors[i],
                              fontsize=12,
                              label_type='edge',
                              fontweight='bold')            
            else: 
                axes[i].bar_label(axes[i].containers[0], 
                              padding=-25,
                              color='white',
                              fontsize=12,
                              label_type='edge',
                              fontweight='bold')
            i += 1            
        plt.show()
        if not in_percents:
            plt.figure(figsize=(5, 5))
            plt.title(x_label) 
            cat_df = df[column].value_counts(normalize=True)
            plt.pie(cat_df,
                    labels=cat_df.index, 
                    autopct='%1.1f%%', 
                    wedgeprops = { 'linewidth': 3, 'edgecolor': 'white' })
            plt.show();
````

````python
# создадим функцию для построения диаграммы корреляции признаков
def corr_matrix(data, columns, interval_cols):
    """
    Отображает матрицу кореляции.

    Параметры:
    - data: Датасет.
    - columns: Названия колонок.
    - interval_cols: Числовые колонки.
    
   
    Пример:
    ```
    corr_matrix(data, columns, interval_cols)
    ```
    """
    
    # сформируем матрицу корреляции
    corr_matrix = data[columns].phik_matrix(interval_cols=interval_cols)
    # построим тепловую карту
    plt.figure(figsize=(18, 12))
    sns.heatmap(corr_matrix, 
                annot=True, 
                linewidths=0.5, 
                cmap='coolwarm')
    plt.show();
````

### Подготовка пайплайна и обучение модели регрессии 

#### Подготовка метрики SMAPE

Оценку качества имодели будем проводить на основе метрики `SMAPE` (англ. symmetric mean absolute percentage error, «симметричное среднее абсолютное процентное отклонение»). 

**Напишем функцию для вычисления SMAPE:**

````python
def smape_get(target, predict):
    return ((1 / len(target)) * np.sum((np.abs(target - predict)) / ((np.abs(target) + np.abs(predict)) / 2)) * 100);
````

Создадим оъект для оценки моделей:

````python
scorer = make_scorer(smape_get, greater_is_better=False);
````

#### Подготовка пайплайна

````python
# Размер исходного датафрейма
len_start = train_job_satisfaction_rate_df.shape[0]
# Создадим новый датафрейм с индексом на основе `id`
train_task_1 = train_job_satisfaction_rate_df.set_index('id')
print(f'Явные дубликаты в train_task_1: {train_task_1.duplicated().sum()}')
train_task_1 = train_task_1.drop_duplicates()
# Проверим результат
display(train_task_1.head())
len_finish = train_task_1.shape[0]
print(f'Размер нового датафрейма {len_finish}')
print(f'Удалено {len_start - len_finish} дубликатов или {round(((len_start - len_finish) * 100 / len_start), 1)}% данных.');
````

````python
# Подготовим выборки
X_train = train_task_1.drop(['job_satisfaction_rate'], axis=1)
y_train = train_task_1['job_satisfaction_rate'];
````

````python
# Подготовим списки признаков для кодирования
ohe_columns = ['dept', 'last_year_promo', 'last_year_violations']
ord_columns = ['level', 'workload']
num_columns = ['salary', 'employment_years', 'supervisor_evaluation'];
````

````python
# Пайплайн для OHE-кодирования
ohe_pipe = Pipeline(
    [
        (
            'simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'simpleImputer_ohe_blank', SimpleImputer(missing_values=' ', strategy='most_frequent')
        ),
        (
            'ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
        )
    ]
);
````

````python
# Пайплайн для ORD-кодирования
ord_pipe = Pipeline(
    [
        (
            'simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',  OrdinalEncoder(
                categories=[
                    ['junior', 'middle', 'sinior'],
                    ['low', 'medium', 'high'],
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
        (
            'simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
);
````

````python
# Общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
);
````

````python
# Итоговый пайплайн
pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', LinearRegression())
    ]
);
````

````python
# Модели
param_grid = [
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 25),
        'models__max_features': range(2, 12),
        'models__min_samples_split': [2, 5, 10],  
        'models__min_samples_leaf': [1, 2, 4], 
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    {
        'models': [LinearRegression()],
        'models__fit_intercept': [True, False], 
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
];
````

#### Подбор лучшей модели и ее параметров

````python
%%time
randomized_search_1 = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring=scorer,
    random_state=RANDOM_STATE,
    error_score='raise',
    n_jobs=-1
)
randomized_search_1.fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', randomized_search_1.best_estimator_)
print (f'Метрика SMAPE лучшей модели на тренировочной выборке: {round(randomized_search_1.best_score_ * -1, 2)}%')
````

````python
# Командой randomized_search.cv_results_ получим все результаты и сделаем из них датафрейм.
# Выведем четыре столбца, отсортируем по rank_test_score — рейтингу качества моделей:
result = pd.DataFrame(randomized_search_1.cv_results_)
with pd.option_context('max_colwidth', 3000):
    display(result[
        ['rank_test_score', 'param_models', 'mean_test_score','params']
    ].sort_values('rank_test_score').head(10))
````

#### Проверка модели на тестовой выборке

````python
# Подготовим тестовый датафрейм
test_task_1 = pd.merge(test_features_df, test_target_job_satisfaction_rate_df, on='id', how='left')
# Проверим результат
display(test_task_1.head())
len_finish = test_task_1.shape[0]
print(f'Размер нового датафрейма {len_finish}')
# Подготовим выборки
X_test = test_task_1.drop(['id', 'job_satisfaction_rate'], axis=1)
y_test = test_task_1['job_satisfaction_rate']
# Получим предсказания
y_test_pred = randomized_search_1.predict(X_test)
# Получим иетрику
print(f'Метрика SMAPE на тестовой выборке: {round(smape_get(y_test, y_test_pred), 2)}%');
````

### Промежуточные выводы

**В ходе подбора лучшей модели и её гиперпараметров с использованием RandomizedSearchCV, лучшей была выбрана модель `DecisionTreeRegressor` с оптимальными гиперпараметрами: `max_depth` = 21, `max_features` = 11, `min_samples_leaf` = 2, `min_samples_split` = 10.**

Метрика `SMAPE` одинаково учитывает масштаб и целевого признака, и предсказания, устойчива к выбросам и хорошо интерпретируется. 

**В результате подбора оптимальных параметров удалось достичь значения `SMAPE` = 15.23% для тренировочной выборки и 13.47% для тестовой. Заданный критерий успеха `SMAPE` ≤ 15 на тестовой выборке достигнут.**

### Подготовка пайплайна и обучение модели классификации

#### Подготовка пайплайна

````python
# Размер исходного датафрейма
len_start =  train_quit_df.shape[0]
# Создадим новый датафрейм с индексом на основе `id`
train_task_2 = train_quit_df.set_index('id')
print(f'Явные дубликаты в train_task_2: {train_task_2.duplicated().sum()}')
train_task_2 = train_task_2.drop_duplicates()
# Проверим результат
display(train_task_1.head())
len_finish = train_task_2.shape[0]
print(f'Размер нового датафрейма {len_finish}')
print(f'Удалено {len_start - len_finish} дубликатов или {round(((len_start - len_finish) * 100 / len_start), 1)}% данных.');
````

````python
# Подготовим выборки
X_train = train_task_2.drop(['quit'], axis=1)
y_train = train_task_2['quit']
````

````python
# Подготовим списки признаков для кодирования
ohe_columns = ['dept', 'last_year_promo', 'last_year_violations']
ord_columns = ['level', 'workload']
num_columns = ['salary', 'employment_years', 'supervisor_evaluation', 'job_satisfaction_rate'];
````

````python
# Пайплайн для OHE-кодирования
ohe_pipe = Pipeline(
    [
        (
            'simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'simpleImputer_ohe_blank', SimpleImputer(missing_values=' ', strategy='most_frequent')
        ),
        (
            'ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
        )
    ]
);
````

````python
ord_pipe = Pipeline(
    [
        (
            'simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',  OrdinalEncoder(
                categories=[
                    ['junior', 'middle', 'sinior'],
                    ['low', 'medium', 'high'],
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
        (
            'simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
);
````

````python
# Общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
);
````

````python
# Итоговый пайплайн
pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', LinearRegression())
    ]
);
````

````python
# Модели
param_grid = [
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__min_samples_split': range(2, 22),
        'models__min_samples_leaf': range(1, 22),
        'models__max_depth': range(2, 22),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']    },
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2,8),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear', 
            penalty='l1')],
        'models__C': range(1,8),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
];
````

#### Подбор лучшей модели и ее параметров

````python
randomized_search_2 = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search_2.fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', randomized_search_2.best_estimator_)
print (f'Метрика ROC-AUC лучшей модели на тренировочной выборке:{round(randomized_search_2.best_score_, 2)}');
````

````python
# Командой randomized_search.cv_results_ получим все результаты и сделаем из них датафрейм.
# Выведем четыре столбца, отсортируем по rank_test_score — рейтингу качества моделей:
result = pd.DataFrame(randomized_search_2.cv_results_)
with pd.option_context('max_colwidth', 3000):
    display(result[
        ['rank_test_score', 'param_models', 'mean_test_score','params']
    ].sort_values('rank_test_score').head(10));''
````
