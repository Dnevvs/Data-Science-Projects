````python
def product_prediction(df, target):
    """
    Функция доя обучения м валидации модели регрессии для прогнозирования добычи нефти.

    Параметры:
    - df: датасет с исходными данными.
    - target: целевой признак
    Пример:
    ```
        product_prediction(df, 'product')
    ```
    Возвращает словарь с параметрами модели и показателями качества:
    """  
    # Формируем данные для модели
    X = df.drop(target, axis=1)
    y = df[target]
    # Формируем обучающую и валидационную выборки 
    X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.25, random_state=RANDOM_STATE)
    # Список числовых признаков для кодирования
    num_columns = X.select_dtypes(include='number').columns.tolist()
    # Пайплайн для подготовки данных
    data_preprocessor = ColumnTransformer(
        [
            ('num', MinMaxScaler(), num_columns)
        ],
        remainder='passthrough'
    )
    # Итоговый пайплайн
    pipe_final = Pipeline(
        [
            ('preprocessor', data_preprocessor),
            ('models', LinearRegression())
        ]
    )
    # Параметры модели
    param_grid = {
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']   
    }
    # Подбор параметров
    gs = GridSearchCV(
        pipe_final, 
        param_grid, 
        scoring='neg_root_mean_squared_error',
        cv=5,
        n_jobs=-1
    )
    # Обучение модели
    gs.fit(X_train, y_train)
    # Формирование результатов
    result = {}
    result['best_model'] = gs.best_estimator_
    
    # Предсказания на обучающей и валтдационной выборке
    y_pred_valid = gs.predict(X_valid)
    y_pred_train = result['best_model'].predict(X_train)

    result['rmse_valid'] = root_mean_squared_error(y_valid, y_pred_valid)
    result['rmse_train'] = gs.best_score_ * -1
    result['r2_valid'] = r2_score(y_valid, y_pred_valid)
    result['r2_train'] = r2_score(y_train, y_pred_train)
    # Объединим истинные и предсказанные значения целевого признака
    result[target] = pd.DataFrame(y_valid)
    result[target]['prediction'] = y_pred_valid
    return result;
````

````python
def df_research_method(df, method):
    """
    Применяет определенный метод изучения данных в датасете.

    Параметры:
    - df: Исследуемый датасет.
    - method: Название метода.
   
    Пример:
    ```
    df_research_method(df, 'head')
    ```
    """
    
    if method == 'head':
        display(df.head())
    elif method == 'shape':
        display(df.shape)
    elif method == 'describe':
        display(df.describe().T)
    elif method == 'duplicated':
        dp = df.duplicated().sum()
        display(dp)
        if dp > 0:
            display(df[df.duplicated()])
    elif method == 'unique':
        for column in df.select_dtypes(include='object').columns:
            print(f'Уникальные значения в колонке {column}: {df[column].sort_values().unique()}')
    elif method == 'msno':
        display(msno.bar(df))
        plt.show()
    else:
        display(df.info())

# Функция применения методов к списку датасетов
def df_research(df_list, df_methods):
    """
    Применяет метод из списка для изучения данных в списке датасетов.

    Параметры:
    - df_list: Список исследуемых датасетов (имя переменной, описание, имя файла).
    - df_method: Список названий методов.
   
    Пример:
    ```
    df_research(df_list, df_methods)
    ```
    """
    
    for method in df_methods:
        print(f'Изучим {method[1]}:')
        print('-' * 70)
        for df in df_list:
            globals()[df[0]] = globals()[df[0]].applymap(lambda x: x.lower() if isinstance(x, str) else x)
            print('Dataset:', df[1])
            print('-' * 70)
            df_research_method(globals()[df[0]], method[0])
            print('-' * 70);
````

````python
# Функция для отображения количественного признака    
def show_numeric_feature(df, column, title, x_label, y_label):
    """
    Отображает гистограмму и "ящик с усами" для количественного признака.

    Параметры:
    - df: Датасет.
    - column: Название колонки признака.
    - title: Заголовок диаграммы.
    - x_label: Название оси X.
    - y_label: Название оси Y.
   
    Пример:
    ```
    show_numeric_feature(market_file_df, 
                         'выручка',
                         'Выручка за период',
                         'Выручка',
                         'Количество покупателей')
    ```
    """
    
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))
    plt.suptitle(title.upper(), fontsize=16, ha='center')
    sns.histplot(data=df[column], color='g', kde=False, ax=axes[0]) 
    axes[0].set_xlabel(x_label)
    axes[0].set_ylabel(y_label)
    sns.boxplot(data=df[column], color='c')
    axes[1].set_ylabel(x_label)
    plt.show()
    print('Описание признака:', x_label)
    display(df[column].describe().T)
    

# Функция для отображения количественного признака по списку
def show_numeric_feature_by_list(df, features):
    """
    Отображает гистограмму и "ящик с усами" для количественного признака
    по списку.

    Параметры:
    - df: Датасет.
    - features: Список признаков для отображения.
   
    Пример:
    ```
    show_numeric_feature_by_list(market_file_df, 
        market_file_numeric_features)
    ```
    """
    
    for feature in features:
        title = feature[2]
        column = feature[0]
        x_label = feature[1]
        y_label = 'Количество'
        show_numeric_feature(df, column, title, x_label, y_label);
````

````python
def show_barplot(title, x, y, x_label, y_label):
    ax = sns.barplot(x=x, y=y) 
    plt.suptitle(title, fontsize=16, ha='center')
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.bar_label(ax.containers[0],
                 padding=-25,
                 color='white',
                 fontsize=12,
                 label_type='edge',
                 fontweight='bold')
    plt.show();
````

````python
# создадим функцию для построения диаграммы корреляции признаков
def corr_matrix(data, columns, interval_cols):
    """
    Отображает матрицу кореляции.

    Параметры:
    - data: Датасет.
    - columns: Названия колонок.
    - interval_cols: Числовые колонки.
    
   
    Пример:
    ```
    corr_matrix(data, columns, interval_cols)
    ```
    """
    
    # сформируем матрицу корреляции
    corr_matrix = data[columns].phik_matrix(interval_cols=interval_cols)
    # построим тепловую карту
    plt.figure(figsize=(18, 12))
    sns.heatmap(corr_matrix, 
                annot=True, 
                linewidths=0.5, 
                cmap='coolwarm')
    plt.show();
````

### Преобразование колонок

### Функция для расчета прибыли

````python
def profit_calc(target, probabilities):

    target = target.reset_index(drop=True)
    probabilities = probabilities.reset_index(drop=True)
    
    probs_sorted = probabilities.sort_values(ascending=False)[:BEST_REGION_WELLS]
    selected = target[probs_sorted.index]
    return selected.sum() *  INCOME_PER_BURREL - WELL_MINING_BUDGET;
````

### Применение Bootstrap для нахождения  распределения прибыли.

````python
data = []
for i in range(len(datasets)):
    df = datasets[i]
    values = []
    for _ in range(1000):
        target_subsample = results[i][target_feature][target_feature].sample(REGION_WELLS, 
                                                                              replace=True, random_state=STATE) 
        probs_subsample = results[i][target_feature]['prediction'][target_subsample.index]
        values.append(profit_calc(target_subsample, probs_subsample))
    values = pd.Series(values)
    mean = round(values.mean() / 1_000_000, 3)
    # confidence_interval = st.t.interval(0.95, len(values)-1, values.mean(), values.sem())    
    # lower = round(confidence_interval[0] / 1_000_000, 3)
    # upper = round(confidence_interval[1] / 1_000_000, 3)
    lower = round(values.quantile(0.025) / 1_000_000, 3)
    upper = round( values.quantile(0.975) / 1_000_000, 3)
    risk = round(float(100 * (values.apply(lambda x: x < 0).sum()) / len(values)), 1)
    data.append(['Регион '+str(i+1), mean, risk])
    print(f'Регион {i+1}:')
    print(f'Средняя прибыль от 200 лучших скважин: {mean} млн. руб.')
    print(f'95%-й доверительный интервал: от {lower} до {upper} млн. руб.')
    print(f'Риск: {risk}%')
    print('-' * 70)
````
