````python
def product_prediction(df, target):
    """
    –§—É–Ω–∫—Ü–∏—è –¥–æ—è –æ–±—É—á–µ–Ω–∏—è –º –≤–∞–ª–∏–¥–∞—Ü–∏–∏ –º–æ–¥–µ–ª–∏ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –¥–ª—è –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è –¥–æ–±—ã—á–∏ –Ω–µ—Ñ—Ç–∏.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - df: –¥–∞—Ç–∞—Å–µ—Ç —Å –∏—Å—Ö–æ–¥–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏.
    - target: —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫
    –ü—Ä–∏–º–µ—Ä:
    ```
        product_prediction(df, 'product')
    ```
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ª–æ–≤–∞—Ä—å —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ –º–æ–¥–µ–ª–∏ –∏ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è–º–∏ –∫–∞—á–µ—Å—Ç–≤–∞:
    """  
    # –§–æ—Ä–º–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è –º–æ–¥–µ–ª–∏
    X = df.drop(target, axis=1)
    y = df[target]
    # –§–æ—Ä–º–∏—Ä—É–µ–º –æ–±—É—á–∞—é—â—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫–∏ 
    X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.25, random_state=RANDOM_STATE)
    # –°–ø–∏—Å–æ–∫ —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è
    num_columns = X.select_dtypes(include='number').columns.tolist()
    # –ü–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö
    data_preprocessor = ColumnTransformer(
        [
            ('num', MinMaxScaler(), num_columns)
        ],
        remainder='passthrough'
    )
    # –ò—Ç–æ–≥–æ–≤—ã–π –ø–∞–π–ø–ª–∞–π–Ω
    pipe_final = Pipeline(
        [
            ('preprocessor', data_preprocessor),
            ('models', LinearRegression())
        ]
    )
    # –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –º–æ–¥–µ–ª–∏
    param_grid = {
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']   
    }
    # –ü–æ–¥–±–æ—Ä –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
    gs = GridSearchCV(
        pipe_final, 
        param_grid, 
        scoring='neg_root_mean_squared_error',
        cv=5,
        n_jobs=-1
    )
    # –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
    gs.fit(X_train, y_train)
    # –§–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    result = {}
    result['best_model'] = gs.best_estimator_
    
    # –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –∏ –≤–∞–ª—Ç–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
    y_pred_valid = gs.predict(X_valid)
    y_pred_train = result['best_model'].predict(X_train)

    result['rmse_valid'] = root_mean_squared_error(y_valid, y_pred_valid)
    result['rmse_train'] = gs.best_score_ * -1
    result['r2_valid'] = r2_score(y_valid, y_pred_valid)
    result['r2_train'] = r2_score(y_train, y_pred_train)
    # –û–±—ä–µ–¥–∏–Ω–∏–º –∏—Å—Ç–∏–Ω–Ω—ã–µ –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞
    result[target] = pd.DataFrame(y_valid)
    result[target]['prediction'] = y_pred_valid
    return result;
````

````python
def df_research_method(df, method):
    """
    –ü—Ä–∏–º–µ–Ω—è–µ—Ç –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–π –º–µ—Ç–æ–¥ –∏–∑—É—á–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - df: –ò—Å—Å–ª–µ–¥—É–µ–º—ã–π –¥–∞—Ç–∞—Å–µ—Ç.
    - method: –ù–∞–∑–≤–∞–Ω–∏–µ –º–µ—Ç–æ–¥–∞.
   
    –ü—Ä–∏–º–µ—Ä:
    ```
    df_research_method(df, 'head')
    ```
    """
    
    if method == 'head':
        display(df.head())
    elif method == 'shape':
        display(df.shape)
    elif method == 'describe':
        display(df.describe().T)
    elif method == 'duplicated':
        dp = df.duplicated().sum()
        display(dp)
        if dp > 0:
            display(df[df.duplicated()])
    elif method == 'unique':
        for column in df.select_dtypes(include='object').columns:
            print(f'–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –∫–æ–ª–æ–Ω–∫–µ {column}: {df[column].sort_values().unique()}')
    elif method == 'msno':
        display(msno.bar(df))
        plt.show()
    else:
        display(df.info())

# –§—É–Ω–∫—Ü–∏—è –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è –º–µ—Ç–æ–¥–æ–≤ –∫ —Å–ø–∏—Å–∫—É –¥–∞—Ç–∞—Å–µ—Ç–æ–≤
def df_research(df_list, df_methods):
    """
    –ü—Ä–∏–º–µ–Ω—è–µ—Ç –º–µ—Ç–æ–¥ –∏–∑ —Å–ø–∏—Å–∫–∞ –¥–ª—è –∏–∑—É—á–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –≤ —Å–ø–∏—Å–∫–µ –¥–∞—Ç–∞—Å–µ—Ç–æ–≤.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - df_list: –°–ø–∏—Å–æ–∫ –∏—Å—Å–ª–µ–¥—É–µ–º—ã—Ö –¥–∞—Ç–∞—Å–µ—Ç–æ–≤ (–∏–º—è –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π, –æ–ø–∏—Å–∞–Ω–∏–µ, –∏–º—è —Ñ–∞–π–ª–∞).
    - df_method: –°–ø–∏—Å–æ–∫ –Ω–∞–∑–≤–∞–Ω–∏–π –º–µ—Ç–æ–¥–æ–≤.
   
    –ü—Ä–∏–º–µ—Ä:
    ```
    df_research(df_list, df_methods)
    ```
    """
    
    for method in df_methods:
        print(f'–ò–∑—É—á–∏–º {method[1]}:')
        print('-' * 70)
        for df in df_list:
            globals()[df[0]] = globals()[df[0]].applymap(lambda x: x.lower() if isinstance(x, str) else x)
            print('Dataset:', df[1])
            print('-' * 70)
            df_research_method(globals()[df[0]], method[0])
            print('-' * 70);
````

````python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞    
def show_numeric_feature(df, column, title, x_label, y_label):
    """
    –û—Ç–æ–±—Ä–∞–∂–∞–µ—Ç –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –∏ "—è—â–∏–∫ —Å —É—Å–∞–º–∏" –¥–ª—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - df: –î–∞—Ç–∞—Å–µ—Ç.
    - column: –ù–∞–∑–≤–∞–Ω–∏–µ –∫–æ–ª–æ–Ω–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞.
    - title: –ó–∞–≥–æ–ª–æ–≤–æ–∫ –¥–∏–∞–≥—Ä–∞–º–º—ã.
    - x_label: –ù–∞–∑–≤–∞–Ω–∏–µ –æ—Å–∏ X.
    - y_label: –ù–∞–∑–≤–∞–Ω–∏–µ –æ—Å–∏ Y.
   
    –ü—Ä–∏–º–µ—Ä:
    ```
    show_numeric_feature(market_file_df, 
                         '–≤—ã—Ä—É—á–∫–∞',
                         '–í—ã—Ä—É—á–∫–∞ –∑–∞ –ø–µ—Ä–∏–æ–¥',
                         '–í—ã—Ä—É—á–∫–∞',
                         '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π')
    ```
    """
    
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))
    plt.suptitle(title.upper(), fontsize=16, ha='center')
    sns.histplot(data=df[column], color='g', kde=False, ax=axes[0]) 
    axes[0].set_xlabel(x_label)
    axes[0].set_ylabel(y_label)
    sns.boxplot(data=df[column], color='c')
    axes[1].set_ylabel(x_label)
    plt.show()
    print('–û–ø–∏—Å–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∞:', x_label)
    display(df[column].describe().T)
    

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –ø–æ —Å–ø–∏—Å–∫—É
def show_numeric_feature_by_list(df, features):
    """
    –û—Ç–æ–±—Ä–∞–∂–∞–µ—Ç –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –∏ "—è—â–∏–∫ —Å —É—Å–∞–º–∏" –¥–ª—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞
    –ø–æ —Å–ø–∏—Å–∫—É.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - df: –î–∞—Ç–∞—Å–µ—Ç.
    - features: –°–ø–∏—Å–æ–∫ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è.
   
    –ü—Ä–∏–º–µ—Ä:
    ```
    show_numeric_feature_by_list(market_file_df, 
        market_file_numeric_features)
    ```
    """
    
    for feature in features:
        title = feature[2]
        column = feature[0]
        x_label = feature[1]
        y_label = '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ'
        show_numeric_feature(df, column, title, x_label, y_label);
````

````python
def show_barplot(title, x, y, x_label, y_label):
    ax = sns.barplot(x=x, y=y) 
    plt.suptitle(title, fontsize=16, ha='center')
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.bar_label(ax.containers[0],
                 padding=-25,
                 color='white',
                 fontsize=12,
                 label_type='edge',
                 fontweight='bold')
    plt.show();
````

````python
# —Å–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –¥–∏–∞–≥—Ä–∞–º–º—ã –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
def corr_matrix(data, columns, interval_cols):
    """
    –û—Ç–æ–±—Ä–∞–∂–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É –∫–æ—Ä–µ–ª—è—Ü–∏–∏.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - data: –î–∞—Ç–∞—Å–µ—Ç.
    - columns: –ù–∞–∑–≤–∞–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫.
    - interval_cols: –ß–∏—Å–ª–æ–≤—ã–µ –∫–æ–ª–æ–Ω–∫–∏.
    
   
    –ü—Ä–∏–º–µ—Ä:
    ```
    corr_matrix(data, columns, interval_cols)
    ```
    """
    
    # —Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º –º–∞—Ç—Ä–∏—Ü—É –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏
    corr_matrix = data[columns].phik_matrix(interval_cols=interval_cols)
    # –ø–æ—Å—Ç—Ä–æ–∏–º —Ç–µ–ø–ª–æ–≤—É—é –∫–∞—Ä—Ç—É
    plt.figure(figsize=(18, 12))
    sns.heatmap(corr_matrix, 
                annot=True, 
                linewidths=0.5, 
                cmap='coolwarm')
    plt.show();
````

### –ü—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–µ –≤—ã–≤–æ–¥—ã

–í —Ö–æ–¥–µ –∏–∑—É—á–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –≤—ã—è–≤–∏–ª–∏, —á—Ç–æ:
- –ö–æ–ª–æ–Ω–∫–∏ —Ç–∏–ø–∞ `int64`, `float64` –º–æ–∂–Ω–æ –∑–∞–º–µ–Ω–∏—Ç—å –Ω–∞ `int32`, `float32` —Å —Ü–µ–ª—å—é —ç–∫–æ–Ω–æ–º–∏–∏ –ø–∞–º—è—Ç–∏.
- –Ø–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã –Ω–µ –æ–±–Ω–∞—Ä—É–∂–µ–Ω—ã.
- –ü—Ä–æ–ø—É—Å–∫–∏ –≤ –¥–∞–Ω–Ω—ã—Ö –Ω–µ –æ–±–Ω–∞—Ä—É–∂–µ–Ω—ã.



<div class="alert alert-block alert-success">
<b>–£—Å–ø–µ—Ö:</b> –ü–µ—Ä–≤–∏—á–Ω—ã–π –æ—Å–º–æ—Ç—Ä –ø—Ä–æ–≤–µ–¥–µ–Ω

##  –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö

### –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∫–æ–ª–æ–Ω–æ–∫

–ü—Ä–∏–≤–µ–¥–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫ –∫ —Å—Ç–∞–Ω–¥–∞—Ä—Ç—É "snake_case". 
–ö–æ–ª–æ–Ω–∫–∏ —Ç–∏–ø–∞ `int64`, `float64` –∑–∞–º–µ–Ω–∏–º –Ω–∞ `int32`, `float32` —Å —Ü–µ–ª—å—é —ç–∫–æ–Ω–æ–º–∏–∏ –ø–∞–º—è—Ç–∏.
–ü—Ä–æ–≤–µ—Ä–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç.

````python
for df in datasets:
    print('Dataset:', df[1])
    # –ü—Ä–∏–≤–µ–¥–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫ –∫ —Å—Ç–∞–Ω–¥–∞—Ä—Ç—É "snake_case".
    globals()[df[0]].columns = [x.lower().
                 split('(', 1)[0].split(',', 1)[0].replace(' ', '_').rstrip('_') for x in globals()[df[0]].columns]
    # –ö–æ–ª–æ–Ω–∫–∏ —Ç–∏–ø–∞ `int64`, `float64` –∑–∞–º–µ–Ω–∏–º –Ω–∞ `int32`, `float32` —Å —Ü–µ–ª—å—é —ç–∫–æ–Ω–æ–º–∏–∏ –ø–∞–º—è—Ç–∏.
    for column in globals()[df[0]].columns:
        if globals()[df[0]][column].dtypes == 'int64':
            globals()[df[0]][column] = globals()[df[0]][column].astype('int32')
        if globals()[df[0]][column].dtypes == 'float64':
            globals()[df[0]][column] = globals()[df[0]][column].astype('float32')
    # –ü—Ä–æ–≤–µ—Ä–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç.
    display(globals()[df[0]].info());
````

### –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –Ω–µ—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã

````python
for df in datasets:
    print('Dataset:', df[1])
    start_len = globals()[df[0]].shape[0]
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è –Ω–µ—è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤
    print(f'–ù–µ—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã –≤ id: {globals()[df[0]].duplicated(subset=["id"]).sum()}')
    # –£—Å—Ç—Ä–∞–Ω–µ–Ω–∏–µ –Ω–µ—è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤
    globals()[df[0]] = globals()[df[0]].drop_duplicates(subset=['id'], keep='first').set_index('id')
    end_len = globals()[df[0]].shape[0]
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
    print(f'–£–¥–∞–ª–µ–Ω–æ {round(100 * (start_len - end_len) / start_len,2)}% –¥–∞–Ω–Ω—ã—Ö')
    print(f'–î—É–±–ª–∏–∫–∞—Ç—ã: {globals()[df[0]].duplicated().sum()}')
    print('-' * 70);
````

–í —Ö–æ–¥–µ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö:

- –ö–æ–ª–æ–Ω–∫–∏ —Ç–∏–ø–∞ `int64`, `float64` –∑–∞–º–µ–Ω–∏–ª–∏ –Ω–∞ `int32`, `float32` —Å —Ü–µ–ª—å—é —ç–∫–æ–Ω–æ–º–∏–∏ –ø–∞–º—è—Ç–∏.
- –û–±–Ω–∞—Ä—É–∂–µ–Ω—ã –∏ —É—Å—Ç—Ä–∞–Ω–µ–Ω—ã –Ω–µ—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã –≤ `id` —Å–∫–≤–∞–∂–∏–Ω.

````python
# –°–ø–∏—Å–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞
# –≤ —Ñ–æ—Ä–º–∞—Ç–µ - –ø—Ä–∏–∑–Ω–∞–∫, –Ω–∞–∑–≤–∞–Ω–∏–µ, –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π
df_numeric_features = [
    ['id', 'ID —Å–∫–≤–∞–∂–∏–Ω—ã', '—É–Ω–∏–∫–∞–ª—å–Ω—ã–π –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä —Å–∫–≤–∞–∂–∏–Ω—ã'],
    ['f0', '–ü—Ä–∏–∑–Ω–∞–∫ 1', '–ø—Ä–∏–∑–Ω–∞–∫ 1 —Å–∫–≤–∞–∂–∏–Ω—ã'],
    ['f1', '–ü—Ä–∏–∑–Ω–∞–∫ 2', '–ø—Ä–∏–∑–Ω–∞–∫ 2 —Å–∫–≤–∞–∂–∏–Ω—ã'],
    ['f2', '–ü—Ä–∏–∑–Ω–∞–∫ 3', '–ø—Ä–∏–∑–Ω–∞–∫ 3 —Å–∫–≤–∞–∂–∏–Ω—ã'],
    ['product', '–û–±—ä—ë–º –∑–∞–ø–∞—Å–æ–≤, —Ç—ã—Å.–±–∞—Ä—Ä–µ–ª–µ–π', '–æ–±—ä—ë–º –∑–∞–ø–∞—Å–æ–≤ –≤ —Å–∫–≤–∞–∂–∏–Ω–µ (—Ç—ã—Å. –±–∞—Ä—Ä–µ–ª–µ–π)']
]

# –°–ø–∏—Å–æ–∫ —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–∞—Ç–∞—Å–µ—Ç–æ–≤
df_num_columns = globals()[datasets[0][0]].select_dtypes(include='number').columns.tolist();
````

### –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –≤—Å–µ—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.

````python
# –ê–Ω–∞–ª–∏–∑ —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–∞—Ç–∞—Å–µ—Ç–æ–≤
for df in datasets:
    print('Dataset:', df[1])
    show_numeric_feature_by_list(globals()[df[0]], df_numeric_features[1:]);
````

## –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –∫ —Ä–∞—Å—á—ë—Ç—É –ø—Ä–∏–±—ã–ª–∏

### –†–∞—Å—Å—á–µ—Ç –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø—Ä–æ–¥—É–∫—Ç–∞ –≤ –º–µ—Å—Ç–æ—Ä–æ–∂–¥–µ–Ω–∏—è—Ö —Ä–µ–≥–∏–æ–Ω–∞, –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–≥–æ –¥–ª—è —Ä–∞–∑—Ä–∞–±–æ—Ç–∫–∏

````python
# –û–ø—Ä–µ–¥–µ–ª–∏–º –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ—Ñ—Ç–∏ –¥–ª—è –±–µ–∑—É–±—ã—Ç–æ—á–Ω–æ–π —Ä–∞–∑—Ä–∞–±–æ—Ç–∫–∏
min_barrels = WELL_MINING_BUDGET / (BEST_REGION_WELLS * INCOME_PER_BURREL)
print(f'–ú–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –æ–±—ä—ë–º –Ω–µ—Ñ—Ç–∏ –¥–ª—è –±–µ–∑—É–±—ã—Ç–æ—á–Ω–æ–π —Ä–∞–∑—Ä–∞–±–æ—Ç–∫–∏: {round(min_barrels, 2)} —Ç—ã—Å. –±–∞—Ä—Ä–µ–ª–µ–π');
````

### –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –æ–±—ä—ë–º–∞ –Ω–µ—Ñ—Ç–∏ –¥–ª—è –±–µ–∑—É–±—ã—Ç–æ—á–Ω–æ–π —Ä–∞–∑—Ä–∞–±–æ—Ç–∫–∏ –∏ —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–∞–ø–∞—Å–∞ –Ω–µ—Ñ—Ç–∏ –≤ –∫–∞–∂–¥–æ–º —Ä–µ–≥–∏–æ–Ω–µ

````python
best_share = BEST_REGION_WELLS / REGION_WELLS
print(f'–ù–µ–æ–±—Ö–æ–¥–∏–º–∞—è –¥–æ–ª—è —Å–∫–≤–∞–∂–∏–Ω —Å –¥–æ—Å—Ç–∞—Ç–æ—á–Ω—ã–º–∏ –∑–∞–ø–∞—Å–∞–º–∏: {round(best_share * 100, 2)}%');
````

````python
data = []
for i in range(len(datasets)):
    df = datasets[i]
    print(f'–†–µ–≥–∏–æ–Ω {i+1}')
    print('-' * 70)
    mean_product = float(globals()[df[0]]['product'].mean())
    print(f'–°—Ä–µ–¥–Ω–∏–π –∑–∞–ø–∞—Å –Ω–µ—Ñ—Ç–∏: {round(mean_product, 1)} —Ç—ã—Å. –±–∞—Ä—Ä–µ–ª–µ–π')
    well_share = globals()[df[0]].query('product >= @min_barrels').shape[0] / globals()[df[0]].shape[0]
    data.append(['–†–µ–≥–∏–æ–Ω '+str(i+1), round(mean_product, 1), round(well_share * 100, 1)])
    print(f'–î–æ–ª—è —Å–∫–≤–∞–∂–∏–Ω —Å –¥–æ—Å—Ç–∞—Ç–æ—á–Ω—ã–º–∏ –∑–∞–ø–∞—Å–∞–º–∏: {round(well_share * 100, 1)}%');
    print('-' * 70);
````

````python
# C—Ñ–æ—Ä–º–∏—Ä—É–µ–º —Ç–∞–±–ª–∏—Ü—É —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏ —Ç –ø–æ–∫–∞–∂–µ–º –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
columns = [
    'region',
    'mean_product',
    'well_share'
]
region_df = pd.DataFrame(data, columns=columns);
````

````python
# –ü–æ–∫–∞–∂–µ–º –¥–∏–∞–≥—Ä–∞–º–º—É —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Ä–µ–≥–∏–æ–Ω–æ–≤
show_barplot(
    '–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–µ–≥–∏–æ–Ω–æ–≤ –ø–æ –∑–∞–ø–∞—Å–∞–º',
    region_df['region'],
    region_df['mean_product'],
    '',
    '–°—Ä–µ–¥–Ω–∏–µ –∑–∞–ø–∞—Å—ã, —Ç—ã—Å. –±–∞—Ä—Ä–µ–ª–µ–π'
);
````

````python
# –ü–æ–∫–∞–∂–µ–º –¥–∏–∞–≥—Ä–∞–º–º—É —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Ä–µ–≥–∏–æ–Ω–æ–≤
show_barplot(
    '–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–µ–≥–∏–æ–Ω–æ–≤ –ø–æ –¥–æ–ª–µ —Å–∫–≤–∞–∂–∏–Ω —Å –¥–æ—Å—Ç–∞—Ç–æ—á–Ω—ã–º–∏ –∑–∞–ø–∞—Å–∞–º–∏',
    region_df['region'],
    region_df['well_share'],
    '',
    '–î–æ–ª—è —Å–∫–≤–∞–∂–∏–Ω, %'
);
````

### –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –ø—Ä–∏–±—ã–ª–∏

````python
def profit_calc(target, probabilities):

    target = target.reset_index(drop=True)
    probabilities = probabilities.reset_index(drop=True)
    
    probs_sorted = probabilities.sort_values(ascending=False)[:BEST_REGION_WELLS]
    selected = target[probs_sorted.index]
    return selected.sum() *  INCOME_PER_BURREL - WELL_MINING_BUDGET;
````
